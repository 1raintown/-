## 去除干扰信息+显示词云函数

```python
def show_cloud_words(p_type,stopwords,savedir):
    x=p_type['报警内容']+'。'+p_type['出警情况（反馈内容）']
    x_doc = x.apply(lambda x: re.sub("[0-9]+|辅警|报警|民警|年|月|日|时|分|、|：|:|[*]|（.*?）|{.*?}|[.*?]|【.*?】|\n", "",x))
    t=''
    for i in x_doc:
        t=t+i
    text= jieba.cut(t)
    mytext_list=[]
    for seg in text:
        if seg not in stopwords and seg!=" " and len(seg)!=1:
            mytext_list.append(seg.replace(" ",""))
    cloud_text=" ".join(mytext_list) 


    #print(text_0)
    wordcloud= WordCloud(width=2000,height=2000,max_words=30,font_step=1,font_path="SimHei.ttf",background_color='white').generate(cloud_text)

    %pylab inline  
    #如果不适用这句 魔法命令 图形就会另外打开一个窗口 使用这句命令 就会让作图出现在 jupyter notebook的浏览器中
    #plt.title("求助词云")

    plt.imshow(wordcloud,interpolation='bilinear')
    plt.axis('off')
    plt.savefig(savedir)
    plt.show()
```
